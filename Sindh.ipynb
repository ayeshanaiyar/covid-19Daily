```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from sklearn.model_selection import train_test_split
pd.options.mode.chained_assignment = None

#model for testings
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.svm import SVR

import os
if not os.path.exists('Output'):
    os.makedirs('Output')


# Import pandas
import pandas as pd

# Load Excel file into a DataFrame
file_path = 'COVID-19 Daily.xlsx'  
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
print(data.head())

# Load a specific sheet by name
data = pd.read_excel(file_path, sheet_name='Sindh')

# Load a specific sheet by index 
data = pd.read_excel(file_path, sheet_name=0)


# Check for null values
print(data.isnull().sum())

# Display column info
print(data.info())

# Show descriptive statistics
print(data.describe())


import pandas as pd

# Step 1: Load the dataset
file_path = 'COVID-19 Daily.xlsx'  
data = pd.read_excel(file_path)

# Step 2: Convert 'date' column to datetime format if it's not already
data['Date'] = pd.to_datetime(data['Date'])

# Set the 'date' column as index for resampling
data.set_index('Date', inplace=True)

# Step 3: Resample data to weekly and monthly frequency
# For weekly data (resample by 'W' for week-ending Sundays)
weekly_data = data.resample('W').sum()

# For monthly data (resample by 'M' for month-end)
monthly_data = data.resample('M').sum()

# Step 4: Find max and min values for weekly and monthly data
# For Weekly
weekly_max_tests = weekly_data['Daily Tests'].max()
weekly_min_tests = weekly_data['Daily Tests'].min()

weekly_max_cases = weekly_data['Daily Cases'].max()
weekly_min_cases = weekly_data['Daily Cases'].min()

weekly_max_recoveries = weekly_data['Daily Recoveries'].max()
weekly_min_recoveries = weekly_data['Daily Recoveries'].min()

weekly_max_deaths = weekly_data['Daily Deaths'].max()
weekly_min_deaths = weekly_data['Daily Deaths'].min()

# For Monthly
monthly_max_tests = monthly_data['Daily Tests'].max()
monthly_min_tests = monthly_data['Daily Tests'].min()

monthly_max_cases = monthly_data['Daily Cases'].max()
monthly_min_cases = monthly_data['Daily Cases'].min()

monthly_max_recoveries = monthly_data['Daily Recoveries'].max()
monthly_min_recoveries = monthly_data['Daily Recoveries'].min()

monthly_max_deaths = monthly_data['Daily Deaths'].max()
monthly_min_deaths = monthly_data['Daily Deaths'].min()

# Step 5: Find corresponding dates for max/min values
# Weekly dates
weekly_max_tests_date = weekly_data['Daily Tests'].idxmax()
weekly_min_tests_date = weekly_data['Daily Tests'].idxmin()

weekly_max_cases_date = weekly_data['Daily Cases']


print(data.columns)


import pandas as pd

# Step 1: Load the dataset
file_path = 'COVID-19 Daily.xlsx'  
data = pd.read_excel(file_path)

# Step 2: Convert 'date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Step 3: Filter the data to include only rows until September 1, 2020
cutoff_date = '2020-09-01'
filtered_data = data[data['Date'] <= cutoff_date]

# Step 4: Sum the 'tests' column for the filtered data
total_tests_till_sep_01 = filtered_data['Daily Tests'].sum()

# Print the result
print(f"Total tests performed until September 01, 2020: {total_tests_till_sep_01}")





plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Daily Tests'], color='blue', label='Daily Tests')
plt.title('Daily Testing Trend Line Graph')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()



plt.figure(figsize=(10, 6))
plt.scatter(data['Date'], data['Daily Tests'], color='red', label='Tests')
plt.title('Daily Testing Scatter Diagram')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Daily Tests'], color='green')
plt.title('Daily Testing Bar Graph')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = 'COVID-19 Daily.xlsx'  
data = pd.read_excel(file_path)

# Convert 'date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Step 2: Calculate the average number of daily tests
average_tests = data['Daily Tests'].mean()

# Step 3: Classify tests as above or below average
above_avg = data[data['Daily Tests'] > average_tests]
below_avg = data[data['Daily Tests'] <= average_tests]

# Step 4: Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot above-average tests (in red)
plt.scatter(above_avg['Date'], above_avg['Daily Tests'], color='red', label='Above Average')

# Plot below-average tests (in green)
plt.scatter(below_avg['Date'], below_avg['Daily Tests'], color='green', label='Below Average')

# Add title and labels
plt.title('Scatter Diagram of Above and Below-Average Daily Testing')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()


import pandas as pd

# Load the dataset
file_path = 'COVID-19 Daily.xlsx'  
data = pd.read_excel(file_path)

# Convert 'date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the 'date' column as index for resampling
data.set_index('Date', inplace=True)

# Resample data to weekly and monthly frequency
weekly_data = data.resample('W').sum()
monthly_data = data.resample('M').sum()

# Find max and min values for weekly data
weekly_max_tests = weekly_data['Daily Tests'].max()
weekly_min_tests = weekly_data['Daily Tests'].min()
weekly_max_cases = weekly_data['Daily Cases'].max()
weekly_min_cases = weekly_data['Daily Cases'].min()
weekly_max_recoveries = weekly_data['Daily Recoveries'].max()
weekly_min_recoveries = weekly_data['Daily Recoveries'].min()
weekly_max_deaths = weekly_data['Daily Deaths'].max()
weekly_min_deaths = weekly_data['Daily Deaths'].min()

# Find corresponding dates for weekly max and min values
weekly_max_tests_date = weekly_data['Daily Tests'].idxmax()
weekly_min_tests_date = weekly_data['Daily Tests'].idxmin()
weekly_max_cases_date = weekly_data['Daily Cases'].idxmax()
weekly_min_cases_date = weekly_data['Daily Cases'].idxmin()
weekly_max_recoveries_date = weekly_data['Daily Recoveries'].idxmax()
weekly_min_recoveries_date = weekly_data['Daily Recoveries'].idxmin()
weekly_max_deaths_date = weekly_data['Daily Deaths'].idxmax()
weekly_min_deaths_date = weekly_data['Daily Deaths'].idxmin()

# Find max and min values for monthly data
monthly_max_tests = monthly_data['Daily Tests'].max()
monthly_min_tests = monthly_data['Daily Tests'].min()
monthly_max_cases = monthly_data['Daily Cases'].max()
monthly_min_cases = monthly_data['Daily Cases'].min()
monthly_max_recoveries = monthly_data['Daily Recoveries'].max()
monthly_min_recoveries = monthly_data['Daily Recoveries'].min()
monthly_max_deaths = monthly_data['Daily Deaths'].max()
monthly_min_deaths = monthly_data['Daily Deaths'].min()

# Find corresponding dates for monthly max and min values
monthly_max_tests_date = monthly_data['Daily Tests'].idxmax()
monthly_min_tests_date = monthly_data['Daily Tests'].idxmin()
monthly_max_cases_date = monthly_data['Daily Cases'].idxmax()
monthly_min_cases_date = monthly_data['Daily Cases'].idxmin()
monthly_max_recoveries_date = monthly_data['Daily Recoveries'].idxmax()
monthly_min_recoveries_date = monthly_data['Daily Recoveries'].idxmin()
monthly_max_deaths_date = monthly_data['Daily Deaths'].idxmax()
monthly_min_deaths_date = monthly_data['Daily Deaths'].idxmin()

# Save weekly and monthly data to Excel files
with pd.ExcelWriter('weekly_monthly_stats.xlsx') as writer:
    weekly_data.to_excel(writer, sheet_name='Weekly Stats')
    monthly_data.to_excel(writer, sheet_name='Monthly Stats')

# Print max and min values with dates for weekly and monthly data
print("Weekly Max and Min:")
print(f"Max weekly tests: {weekly_max_tests} on {weekly_max_tests_date}")
print(f"Min weekly tests: {weekly_min_tests} on {weekly_min_tests_date}")
print(f"Max weekly cases: {weekly_max_cases} on {weekly_max_cases_date}")
print(f"Min weekly cases: {weekly_min_cases} on {weekly_min_cases_date}")
print(f"Max weekly recoveries: {weekly_max_recoveries} on {weekly_max_recoveries_date}")
print(f"Min weekly recoveries: {weekly_min_recoveries} on {weekly_min_recoveries_date}")
print(f"Max weekly deaths: {weekly_max_deaths} on {weekly_max_deaths_date}")
print(f"Min weekly deaths: {weekly_min_deaths} on {weekly_min_deaths_date}")

print("\nMonthly Max and Min:")
print(f"Max monthly tests: {monthly_max_tests} on {monthly_max_tests_date}")
print(f"Min monthly tests: {monthly_min_tests} on {monthly_min_tests_date}")
print(f"Max monthly cases: {monthly_max_cases} on {monthly_max_cases_date}")
print(f"Min monthly cases: {monthly_min_cases} on {monthly_min_cases_date}")
print(f"Max monthly recoveries: {monthly_max_recoveries} on {monthly_max_recoveries_date}")
print(f"Min monthly recoveries: {monthly_min_recoveries} on {monthly_min_recoveries_date}")
print(f"Max monthly deaths: {monthly_max_deaths} on {monthly_max_deaths_date}")
print(f"Min monthly deaths: {monthly_min_deaths} on {monthly_min_deaths_date}")


import pandas as pd

# Load the dataset
file_path = 'COVID-19 Daily.xlsx' 
data = pd.read_excel(file_path)

# Convert 'date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Define the total population (replace with the actual value)
total_population = 1000000  # Replace with the actual population number

# Calculate cases per thousand
data['cases_per_thousand'] = (data['Daily Cases'] / total_population) * 1000

# Calculate cases per million
data['cases_per_million'] = (data['Daily Cases'] / total_population) * 1000


# Display the updated DataFrame with the new column
print(data.head())

# Optional: Save the updated dataset to a new Excel file
data.to_excel('updated_data.xlsx', index=False)


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'COVID-19 Daily.xlsx'  
data = pd.read_excel(file_path)

# Convert 'date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Feature Engineering: Convert dates to ordinal numbers
data['date_ordinal'] = data['Date'].map(pd.Timestamp.toordinal)

# Identify and handle NaN values in target variables
print("Missing values in cases:", data['Daily Cases'].isna().sum())
print("Missing values in deaths:", data['Daily Deaths'].isna().sum())

# Option 1: Drop rows with NaN values in target variables
data = data.dropna(subset=['Daily Cases', 'Daily Deaths'])

# Option 2: Impute NaN values (if applicable)
# Uncomment the following lines if you prefer imputation
# data['cases'].fillna(data['cases'].mean(), inplace=True)
# data['deaths'].fillna(data['deaths'].mean(), inplace=True)

# Define features and target variables
X = data[['date_ordinal']]
y_cases = data['Daily Cases']
y_deaths = data['Daily Deaths']

# Split the data into training and testing sets
X_train, X_test, y_cases_train, y_cases_test = train_test_split(X, y_cases, test_size=0.2, shuffle=False)
X_train, X_test, y_deaths_train, y_deaths_test = train_test_split(X, y_deaths, test_size=0.2, shuffle=False)

# Train and evaluate Linear Regression Model
lin_reg_cases = LinearRegression()
lin_reg_cases.fit(X_train, y_cases_train)
y_cases_pred = lin_reg_cases.predict(X_test)
mse_cases = mean_squared_error(y_cases_test, y_cases_pred)
print(f"Mean Squared Error (Cases) - Linear Regression: {mse_cases}")

lin_reg_deaths = LinearRegression()
lin_reg_deaths.fit(X_train, y_deaths_train)
y_deaths_pred = lin_reg_deaths.predict(X_test)
mse_deaths = mean_squared_error(y_deaths_test, y_deaths_pred)
print(f"Mean Squared Error (Deaths) - Linear Regression: {mse_deaths}")

# Train and evaluate Decision Tree Model
tree_reg_cases = DecisionTreeRegressor()
tree_reg_cases.fit(X_train, y_cases_train)
y_cases_tree_pred = tree_reg_cases.predict(X_test)
mse_cases_tree = mean_squared_error(y_cases_test, y_cases_tree_pred)
print(f"Mean Squared Error (Cases) - Decision Tree: {mse_cases_tree}")

tree_reg_deaths = DecisionTreeRegressor()
tree_reg_deaths.fit(X_train, y_deaths_train)
y_deaths_tree_pred = tree_reg_deaths.predict(X_test)
mse_deaths_tree = mean_squared_error(y_deaths_test, y_deaths_tree_pred)
print(f"Mean Squared Error (Deaths) - Decision Tree: {mse_deaths_tree}")

# Visualize predictions
plt.figure(figsize=(14, 6))

# Linear Regression
plt.subplot(1, 2, 1)
plt.scatter(X_test, y_cases_test, color='blue', label='Actual Cases')
plt.plot(X_test, y_cases_pred, color='red', label='Predicted Cases')
plt.title('Linear Regression - Daily Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_test, y_deaths_test, color='blue', label='Actual Deaths')
plt.plot(X_test, y_deaths_pred, color='red', label='Predicted Deaths')
plt.title('Linear Regression - Daily Deaths')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.legend()

plt.tight_layout()
plt.show()

# Decision Tree Regression
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_test, y_cases_test, color='blue', label='Actual Cases')
plt.scatter(X_test, y_cases_tree_pred, color='red', label='Predicted Cases')
plt.title('Decision Tree - Daily Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_test, y_deaths_test, color='blue', label='Actual Deaths')
plt.scatter(X_test, y_deaths_tree_pred, color='red', label='Predicted Deaths')
plt.title('Decision Tree - Daily Deaths')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.legend()

plt.tight_layout()
plt.show()


```
